```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn.cross_validation import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor, export_graphviz
```

```python
#Get Bird Bone Dataframe
birds = pd.read_csv('bird.csv', sep=',')
new_columns = ['ID', 'HumerusL', 'HumerusD', 'UlnaL','UlnaD', 'FemurL', 'FemurD',
              'TibiotarsusL', 'TibiotarsusD', 'TarsometatarsusL','TarsometatarsusD', 'Type'] 
birds.columns = new_columns

#Drop rows with NaN values
birds = birds.dropna()
```

```python
# mean rentals for each value of "workingday"
birds.groupby('Type').mean()
```

```python
# mean rentals for each value of "hour"
birds.groupby('Type').mean().plot(figsize=(15,15))
```

```python
# hourly rental trend for "workingday=0"
birds[birds.UlnaL > 75].groupby('Type').mean().plot()
```

```python
birds[birds.FemurL < 40].groupby('Type').mean().plot()
```

```python
# create X and y
feature_cols = ['HumerusL', 'UlnaL']
X = birds.HumerusL
y = birds.UlnaL
X=X.reshape(1, -1)
y=y.reshape(1, -1)
```

```python
linreg = LinearRegression()
linreg.fit(X, y)
print(linreg.coef_)
```

```python
scores = cross_val_score(linreg, X, y, cv=4, scoring='mean_squared_error')
```
